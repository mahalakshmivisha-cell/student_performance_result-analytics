```python
import pandas as pd

df = pd.read_csv("student dataset.csv")

df.head()
```

```python
df["Average"] = df[["Engineering Maths",
                    "Engineering Physics",
                    "Engineering Graphics"]].mean(axis=1)

print(df.head())
```

```python
def grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 80:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= 50:
        return "D"
    else:
        return "F"

df["Grade"] = df["Average"].apply(grade)
print(df.head())
```

```python
df["Pass/Fail"] = df["Average"].apply(lambda x: "Pass" if x >= 50 else "Fail")

print(df.head())
```

```python
performance = pd.DataFrame({
    "Subject": ["Engineering Maths", "Engineering Physics", "Engineering Graphics"],
    "Average Marks": [
        df["Engineering Maths"].mean(),
        df["Engineering Physics"].mean(),
        df["Engineering Graphics"].mean()
    ]
})

print(performance)
```

```python
dept_attendance = df.groupby('Department')['Attendance'].mean()

plt.figure(figsize=(8,5))
plt.bar(dept_attendance.index, dept_attendance.values)

plt.title("Department-wise Average Attendance")
plt.xlabel("Department")
plt.ylabel("Average Attendance")

plt.show()
```

```python
import matplotlib.pyplot as plt


subject_avg = df[['Engineering Maths',
                  'Engineering Physics',
                  'Engineering Graphics']].mean()


plt.figure(figsize=(6,6))
plt.pie(subject_avg,
        labels=subject_avg.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Average Marks Distribution by Subject")
plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt

dept_attendance = df.groupby('Department')['Attendance'].mean()

plt.figure(figsize=(8,5))
plt.plot(dept_attendance.index,
         dept_attendance.values,
         marker='o',
         linewidth=2)

plt.title("Department-wise Average Attendance")
plt.xlabel("Department")
plt.ylabel("Average Attendance")
plt.grid(True)

plt.show()
```
